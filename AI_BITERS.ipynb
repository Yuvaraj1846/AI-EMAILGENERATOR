```python
%%writefile app.py
import os
import streamlit as st
import google.generativeai as genai
from st_copy_to_clipboard import st_copy_to_clipboard

# --- CONFIGURE AI MODEL ---
# Set Google API Key
API_KEY = os.getenv("API_KEY")  # Ensure the key is set in your environment variables
# API_KEY = "AIzaSyDD_Lq_o13j-UbNnQ5KKkFp8IgxkgGTjac"
if API_KEY:
    genai.configure(api_key=API_KEY)
else:
    st.error("⚠️ Google API Key is missing. Please set it as an environment variable.")
    st.stop()

# AI Model Configuration
generation_config = {
    "temperature": 0.7,  # Adjusted for more balanced outputs
    "top_p": 0.9,
    "top_k": 40,
    "max_output_tokens": 1024,  # Adjusted for optimal email length
    "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(model_name="gemini-2.0-flash-exp", generation_config=generation_config)

# --- STREAMLIT UI DESIGN ---
st.set_page_config(page_title="AI Email Generator", page_icon="📧", layout="centered")

# Page Header
st.title("📧 MULTI LANGUAGE AI Personalized Email Generator")
st.write("Generate professional, friendly, or casual emails in seconds using AI!")

st.markdown("---")

# --- INPUT FORM ---
with st.form("email_input_form"):
    st.subheader("📋 Provide Email Details")

    col1, col2 = st.columns(2)

    with col1:
        recipient_name = st.text_input("Recipient's Name", placeholder="SHIVA")
        event_name = st.text_input("Event Name", placeholder="NASSCOM HACKATHON")

    with col2:
        event_date = st.text_input("Event Date", placeholder="FEB 1, 2025")
        event_location = st.text_input("Event Location", placeholder="KPRIT, GHANPUR")

    tone = st.selectbox("🎭 Tone of the Email", ["Formal", "Casual", "Friendly", "Professional"])

    language = st.selectbox("🌐 Select Language", ["English", "Hindi", "Telugu"])

    special_instructions = st.text_area("📝 Additional Instructions", placeholder="Mention the early bird registration discount (if needed).")

    # Submit Button
    submit_button = st.form_submit_button("🚀 Generate Email")

# --- GENERATE EMAIL ---
if submit_button:
    if not recipient_name or not event_name or not event_date or not event_location:
        st.warning("⚠️ Please fill in all the required fields.")
    else:
        st.info("⏳ Generating your personalized email...")

        # Adjusting the prompt based on the selected language
        if language == "English":
            prompt = f"""
            Write a {tone.lower()} invitation email for {recipient_name} to attend {event_name} on {event_date} at {event_location}.
            {special_instructions if special_instructions else ''}
            Only return the email content, without any additional commentary, disclaimers, or explanations.
            """
        elif language == "Hindi":
            prompt = f"""
            {recipient_name} को {event_name} में {event_date} को {event_location} में भाग लेने के लिए एक {tone.lower()} निमंत्रण ईमेल लिखें।
            {special_instructions if special_instructions else ''}
            केवल ईमेल सामग्री लौटाएं, अतिरिक्त टिप्पणी, अस्वीकरण या व्याख्या के बिना।
            """
        elif language == "Telugu":
            prompt = f"""
            {recipient_name} గారికి {event_name} లో {event_date} న {event_location} వద్ద హాజరయ్యేందుకు {tone.lower()} ఆహ్వాన ఇమెయిల్ వ్రాయండి.
            {special_instructions if special_instructions else ''}
            కేవలం ఇమెయిల్ విషయాన్ని మాత్రమే తిరిగి ఇవ్వండి, అదనపు వ్యాఖ్యలు, హెచ్చరికలు లేదా వివరణలు లేకుండా.
            """

        # AI Model Interaction
        try:
            response = model.generate_content(prompt)
            generated_email = response.text.strip()

            # Display Generated Email
            st.success("✅ Email Generated Successfully!")
            st.subheader("📨 Generated Email:")
            st_copy_to_clipboard(generated_email)
            email_text_area = st.text_area("Generated Email:", generated_email, height=250)

        except Exception as e:
            st.error(f"❌ Error generating email: {e}")

# --- FOOTER ---
st.markdown("---")
st.caption("🔹 Powered by Google Gemini AI | Created by AI BITERS 🐒")
```
